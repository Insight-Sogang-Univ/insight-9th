# 머신러닝(ML) 개요

머신러닝은 크게 **지도학습, 비지도학습, 강화학습**으로 나뉜다. 

**지도학습**: "정답"을 알고 있는 데이터를 이용해 컴퓨터를 학습시킨다. 이때 컴퓨터는 자신의 답과 정답과 차이를 통해 지속해서 학습한다. 지도학습에는 회귀(Regression), **분류(Classification)** 등 많은 머신러닝 기법들이 해당한다.

**비지도학습**은 지도학습과는 다르게 정답이 있지 않다. 즉, 비지도학습은 정답이 없이 주어진 데이터로만 학습한다. 대표적으로 **군집화(Clustering)**가 비지도학습에 해당한다.

**강화학습**은 지도학습, 비지도학습과 그 성격이 다르다. 강화학습은 **"보상"**을 통해서 학습한다. 즉, 컴퓨터는 컴퓨터가 선택한 행동(action)에 대한 환경에 반응함으로써 학습한다. 이 보상은 직접적인 답은 아니지만 컴퓨터에게는 간접적인 정답의 역할을 한다.

## 머신러닝 중 지도학습을 간단하게 말하면?

머신러닝을 어렵게 생각하시는 분이 많은데 가장 간단하게는 내가 기계(머신)을 학습(러닝) 시킨다고 이해하시면 됩니다. 예를 들어서, 수학을 아무것도 모르는 친구가 고등학교 1학년에 입학을 해서 점점 모의고사를 치고 공부를 하면서 학습을 한다고 상상해보세요. 처음에는 아무것도 모르는 친구가 수업을 듣고, 학원을 다니고, 숙제를 하면서 점점 개념에 대해 익숙해지고 모의고사 점수가 올라가겠죠??   
즉 X(모의고사) 를 학생에게 주입을 해서 Y(모의고사 정답)을 알려주면서 학생을 학습 시켜서 하나의 머신을 만드는 것이 목적입니다. 결과적으로 학습이 잘 된 머신은 기출 말고도 사설 모의고사나 다른 모의고사를 주면 높은 점수를 받겠죠?

# 분류(Classification) 개요

**분류(Classification)**란 기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지한 뒤에 새롭게 관측된 데이터에 대한 레이블을 판별하는 것이다.

분류의 대표적인 머신러닝 알고리즘
- 베이즈(Bayes) 통계와 생성 모델에 기반한 **나이브 베이즈(Naive Bayes)**
- 독립변수와 종속변수의 선형 관계성에 기반한 **로지스틱 회귀(Logistic Regression)**
- 데이터 균일도에 따른 규칙 기반의 **결정 트리(Decision Tree)**
- 개별 클래스 간의 최대 마진 분류 마진을 효과적으로 찾아주는 **서포트 벡터 머신(SVM)**
- 근접 거리를 기준으로 하는 최소 근접(Nearest Neighbor) 알고리즘
- 심층 연결 기반의 신경망(Neural Network)
- 서로 다른(또는 같은) 머신러닝 알고리즘을 결합한 **앙상블(Ensemble)**

## 손실함수

- 손실함수, 비용함수, 목적함수는 의미가 미묘하게 다르다. 하지만 보통 혼용해서 사용한다.
- 굳이 따지자면 손실함수 ≤ 비용함수 ≤ 목적함수

### 손실함수 정의

- 예측값과 실제값의 차이를 구하는 함수
- 이를 통해 오차값을 구하고, 오차값에 기반해서 모델을 학습한다(오차값을 최소화하는 방향으로)
- 한편, 손실함수는 값의 차이를 구하는 함수이므로, **값의 형태**를 구분하는 것이 필수적이다

### 손실함수-분류

#### Entropy: 계량화된 불확실성의 정도
- 정보 엔트로피는 열역학에서의 엔트로피와 크게 다르지 않은 개념으로, 정보 엔트로피(정보량)이 높을수록 불확실성이 커진다.
- 특정 사건이 **발생할 확률이 낮을수록** 어떤 정보인지는 불확실해지고, 이때 ‘**정보가 많다**’, 혹은 ‘**엔트로피가 높다**’고 표현한다.
        
$$Info \propto \frac{1}{P(X)}$$
        
    
- 이때 정보 엔트로피를 나타내는 H(x)는 다음과 같다.
    
$$H(x) = -\sum_{i=1}^{n}(p(x_{i})\cdot log(p(x_{i})))$$
    
- 확률 **p**(0~1 사이)를 특정 사건이 일어날 확률이라고 생각하자.
- 1/p 는 **'놀라움의 정도'**로 볼 수 있다. 특정 사건이 일어날 확률이 작으면(=분모가 작으면), 놀라움의 정도는 커진다.(=1/p는 커진다)
- 각각의 관측 값에 대해 확률이 같다고 가정할 때, 이 식을 정리하면 다음과 같이 나타난다 (ex. 주사위는 1의 눈과 6의 눈이 나올 확률 p가 1/6로 같다)
    
$$p(x_{i}) = p, n=1/p$$    

- n은 데이터의 양(정보량)이다. 앞서 설명하였듯 정보량은 확률에 반비례 하므로, 주어진 식의 계산을 위해 다음과 같이 치환해주면, 다음 식으로 정리된다
    
$$-\sum_{i=1}^{n}(p\cdot log(p))=-\frac{n(n+1)}{2}(p\cdot log(p))=-\frac{p+1}{2p}\times log(p)$$

- 이렇게 정리한 H(x)(정보 엔트로피)를 그래프로 그리면,



0~1 사이의 값 확률 p에서 **0에 가까울수록 양의 무한대로 발산**, **1에 가까울수록 0으로 수렴**하는 것을 볼 수 있다.

**즉, 확률이 0에 가까울수록 엔트로피 값은 커지고, 확률이 1에 가까울수록 엔트로피 값은 0으로 수렴한다.**

#### Cross Entropy
    
$$H_{p}(q) = -\sum_{i=1}^{n}(q(x_{i})\cdot log(p(x_{i})))$$    
- 특정 사건이 일어날 확률(q)을 알지 못하는 상태에서, 표본을 통해 구한 분포(확률) P를 이용해 Q를 예측하는 것
- 크로스 엔트로피가 낮을수록 추정된 클래스의 확률이 타깃 클래스에 잘 맞
- 보통 크로스 엔트로피에 기반해 분류 모형을 학습한다(Target의 class 개수 등에 따라 사용하는 크로스 엔트로피 함수가 달라진다)

### 손실함수 vs 평가지표
- 둘 모두 모형의 성능을 나타내는 지표
- MSE, MAE 등 몇몇 지표는 회귀분석에서 손실함수에 사용되기도, 평가지표로 사용되기도 함
- 손실함수는 Training data를 통해 모델을 학습시키는데 이용
- 평가지표는 Test data, Validation data에서 과대적합, 과소적합이 있는지, 모델의 성능이 좋은지 나쁜지 확인할 때 사용
- 즉, 손실함수가 되기 위해서는 **미분을 통해 파라미터의 값을 조정**할 수 있어야 함

## 평가지표(Metrics)

### 평가지표(회귀)
- MSE, MAE, RMSE 등 Loss function에 사용되는 지표가 사용될 수 있음.
- 그러나 결정계수, 조정된 결정계수, AIC, BIC 등 다른 지표가 사용될 수 있음.

### 평가지표(분류)
#### Confusion Matrix(혼동행렬)
    

    
- TP: 실제 참, 예측 참
- TN: 실제 거짓, 예측 거짓
- FP: 실제 거짓, 예측 참 → 2종 오류(잘못된 귀무가설을 채택했을 때 발생하는 오류)
- FN: 실제 참, 예측 거짓 → 1종 오류(올바른 귀무가설을 기각했을 때 발생하는 오류)

#### Accuracy
$$\frac{TP+TN}{TP+TN+FP+FN}$$
- 가장 익숙한 지표
- Target의 class 분포에 비례하는 경향
    - 10번 중 9번은 동전의 앞면, 1번은 동전의 뒷면이 나온다고 생각하자. 이 경우, 모두 앞면이 나온다고 예측을 해도 정확도는 90%가 나온다
    - 카드사에서 신용카드 연체 여부를 예측하는 모형을 만든다고 생각하자. 이때, 실제로 연체하는 사용자의 비율은 전체의 5%이다. 그렇다면, 모두 연체하지 않는다고 예측해도 정확도는 0.95로 나온다.

#### Precision
$$\frac{TP}{TP+FP}$$ 
- 관건은 FP(실제로 거짓, 예측을 참)
- 실제로 참이고 예측도 참 / 예측을 참
- 거짓을 참으로 판단한 정도

#### Recall
$$\frac{TP}{TP+FN}$$ 
- 관건은 FN(실제로 참, 예측은 거짓)
- 실제로 참이고 예측도 참 / 실제로 참
- 참을 거짓으로 판단한 정도

- 보통 분류를 할 때, 확률에 기반해서 0.5가 넘으면 참, 0.5 미만이면 거짓으로 판단한다. 따라서 이 경곗값(Threshold)를 조정하면 Precision과 Recall을 조정할 수 있다.
- 참으로 후하게 판단하면(Threshold를 낮추면), 거짓을 참으로 판단하는 경우는 많아지고(FP가 많아지고), 참을 거짓으로 판단하는 경우가 적어진다(FN이 줄어든다) 따라서 Precision이 낮아지고, Recall은 높아진다.
- 반대로 참을 엄격하게 판단하면(Threshold를 높이면), 거짓을 참으로 판단하는 경우는 줄어든다(FP는 적어진다). 그러나 참을 거짓으로 판단하는 경우는 늘어난다(FN은 증가한다). 따라서 Precision은 높아지고, Recall은 낮아진다.
- Recall과 Precision은 Trade-off 관계에 있으므로, 분석의 목적에 알맞은 지표를 사용해야 한다
    - 암 환자 판단
        - 암을 검진했을 때, 양성 환자를 음성으로 판단하는 것과 음성 환자를 양성으로 판단할 때를 비교해보자
        - 음성 환자를 양성으로 판단했을 때는 추가 검사를 통해 식별할 수 있다. 그러나 양성 환자를 음성으로 판단했을 때는, 환자가 사망할 수 있다. 즉, 실제 참, 예측 거짓(FN)을 줄여야 한다. 즉, Recall을 높여야 한다. 따라서 Threshold를 낮춘다
    - 스팸 메일 분류
        - 스팸 메일이면 Positive, 정상 메일이면 Negative로 분류하는 스팸 메일 분류기를 생각해보자
        - 일반메일함에 스팸 메일이 오는 것은 큰 문제가 없다. 클릭 몇 번으로 삭제할 수 있기 때문이다. 그러나 스팸 메일함에 일반 메일이 가는 것은 위험하다. 등록금 고지서, 입사 통지서 등을 확인하지 못할 수 있기 때문이다. 따라서, 실제 거짓, 예측 참(FP)를 줄여야 한다. 따라서, Precision을 높여야 하고, Threshold를 높인다.
    - 그러나 무엇을 Positive로 할지 Negative로 할지는 상황에 따라 달라지므로, 무작정 암기보다는 제대로 이해하는 것이 중요하다
    - 스팸메일 분류기에서 정상 메일을 Positive로 스팸 메일을 Negative로 생각한다면, Recall을 높여야한다.

```python
# 이 코드는 Precision과 Recall의 Trade-off 관계를 보여주기 위한 전처리+파일 입출력 코드(중요X)

import urllib.request
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score , recall_score , confusion_matrix
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import Binarizer
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
%matplotlib inline
from sklearn.metrics import precision_recall_curve

# 결측값 대체 함수
def fillna(df):
    df['Age'].fillna(df['Age'].mean(), inplace=True)
    df['Cabin'].fillna('N', inplace=True)
    df['Embarked'].fillna('N', inplace=True)
    df['Fare'].fillna(0, inplace=True)
    return df

# 필요없는 피처 제거
def drop_features(df):
    df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
    return df

# 피처를 라벨 인코딩하기
def format_features(df):
    df['Cabin'] = df['Cabin'].str[:1]
    features = ['Cabin', 'Sex', 'Embarked']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df

#위의 3가지 함수를 한 번에 적용하는 함수
def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)
    return df

url = 'https://raw.githubusercontent.com/wikibook/pymlrev2/main/1%EC%9E%A5/titanic_train.csv'

titanic_df = pd.read_csv(urllib.request.urlopen(url))
y_titanic_df = titanic_df['Survived']
X_titanic_df= titanic_df.drop('Survived', axis=1)
X_titanic_df = transform_features(X_titanic_df)

X_train, X_test, y_train, y_test = train_test_split(X_titanic_df, y_titanic_df, \
                                                    test_size=0.20, random_state=11)

lr_clf = LogisticRegression(solver='liblinear')

lr_clf.fit(X_train , y_train)
```

```python
# Precision과 Recall의 Trade-off 관계를 시각화하기 위한 코드

def precision_recall_curve_plot(y_test, pred_proba_c1):
    precisions, recalls, thresholds = precision_recall_curve(y_test, pred_proba_c1)
    
    plt.figure(figsize=(8,6))
    threshold_boundary = thresholds.shape[0]
    plt.plot(thresholds, precisions[0:threshold_boundary], linestyle='--', label='precision')
    plt.plot(thresholds, recalls[0:threshold_boundary],label='recall')
    
    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1),2))
    
    plt.xlabel('Threshold value'); plt.ylabel('Precision and Recall value')
    plt.legend(); plt.grid()
    plt.show()
    
precision_recall_curve_plot(y_test, lr_clf.predict_proba(X_test)[:, 1] )

# 정밀도와 재현율은 Trade-off 관계임
# 임계값을 조정해 어느 하나의 값을 높일 수 있으므로, 두 값 모두를 고려
```

#### F1-score
    
$$F1\ Score=2\times \frac{recall\times precision}{recall + precision}$$    

- Precision과 Recall을 결합한 수치
- 정밀도와 재현율이 어느 한쪽으로 치우치지 않을 때 좋은 값(큰 값을 가짐)

#### ROC curve
    


- 세로축이 TP/(TP+FN), 가로축이 TN/(TN+FP)
- 모든 Threshold 값에 대해 FPR의 변화에 따른 TPR의 변화를 나타냄
- 따라서 ROC curve의 기울기는 TPR을 증가시키기 위해 감수해야 하는 FPR의 크기
- 최적 Threshold를 정하는 기준점(B와 C는 해당 범주로 예측했을 때의 Benefit, Cost)
        
$$Optimal\ Slope\ of\ ROC=\frac{Pr(neg)}{Pr(pos)} \times \frac{B_{TP}-C_{FP}}{B_{TP}-C_{FN}}$$        
- AUC score: Area Under the Curve, ROC 곡선의 넓이를 나타냄.

```python
# Roc curve를 시각화하기 위한 코드 

from sklearn.metrics import roc_curve

def roc_curve_plot(y_test, pred_proba_c1):
    fprs, tprs, thresholds = roc_curve(y_test, pred_proba_c1)
    plt.plot(fprs, tprs, label = 'ROC')
    plt.plot([0,1], [0,1], 'k--', label = 'Random')

    start, end = plt.xlim()
    plt.xticks(np.round(np.arange(start, end, 0.1), 2))
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('FPR( 1 - Sensitivity )')
    plt.ylabel('TPR( Recall )')
    plt.legend()
    plt.show()

roc_curve_plot(y_test, lr_clf.predict_proba(X_test)[:,1])
```

```python
from sklearn.metrics import roc_auc_score
pred_proba = lr_clf.predict_proba(X_test)[:, 1]
roc_score = roc_auc_score(y_test, pred_proba)
print(f"ROC-AUC 값: {roc_score:.4f}")
```

## 로지스틱 회귀분석
종속변수가 이항분포를 따르는 범주형(Categorical) 변수일 경우 사용하는 분류기법
ex) 공부시간(0~∞시간)과 시험합격여부(탈락 / 합격)

일반선형회귀와 로지스틱 회귀의 차이

### 0. 오즈란?  
- $odds=\frac { P(A) }{ P({ A }^{ c }) } =\frac { P(A) }{ 1-P(A) }$<br><br>


합성함수의 관계(도움이 되길...ㅠ) :   
__증가 o 증가  = 증가__  
__증가 o 감소  = 감소__  
__감소 o 증가  = 감소__  
__감소 o 감소  = 증가__  



### 1. 오즈와 합성함수의 관계

1. 이항(성공 / 실패) → 확률(0 ~ 1) : 로지스틱 함수[Logistic(Sigmoid) Function], Odds(승산)<br>
    - $odds=\frac { P(A) }{ P({ A }^{ c }) } =\frac { P(A) }{ 1-P(A) }$<br><br>
    __로지스틱 회귀분석에서 오즈의 개념이 사용되는 이유 => 0에서 1까지의 값만을 갖는 확률을 실수 전체의 값을 갖는 일반적인 회귀모형에 적용할 수 없으므로, 0에서 무한대의 범위를 갖는 오즈를 사용__<br><br>
    - ${ P(A) } → 0  :  odds → 0$
    - ${P(A)} → 1 : odds → ∞$
    - 종속변수  $odds$의 범위가 $0 \sim ∞$로 바뀌게 된다.
    - 하지만, 독립변수와 달리, 아직 음수에서 정의되지 않는다.

2. **'음의 무한대부터 양의 무한대까지'로 범위 확장하기**
    - $log{(odds)}$
    - $log$를 취하게 되면, 종속변수의 범위가 음의 무한대 부터 양의 무한대로 확장된다.

## SVM

- SVM: 분류를 위한 가장 좋은 결정경계를 찾는 모형
    
    
- 용어 정리
    - Hyperplane: 초평면, 결정경계(N-1 차원)
    - Support Vector: 경계에 있는 data(N+1개)
        - 만약 Support Vector가 N개라면 무수히 많은 결정 경계가 생성됨
    - Margine: Hyperplane과 Support Vector 사이의 거리

    → **Margine을 극대화하는 Hyperplane**을 찾는다

```python
# 시각화를 위한 데이터 읽기와 차원축소 코드(중요X)

!pip install mlxtend
import sklearn.datasets as d
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.decomposition import PCA

x = d.load_breast_cancer()
cancer = pd.DataFrame(data = x.data, columns = x.feature_names)
cancer['target'] = x.target

X = cancer.iloc[:,:-1]
y = cancer.iloc[:,-1]

scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

# 변환된 X로 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.3, random_state = 100)

train_X_reduced = PCA(n_components=2).fit_transform(X_train)
test_X_reduced = PCA(n_components=2).fit_transform(X_test)
X_reduced = PCA(n_components=2).fit_transform(X_scaled)
```

```python
def visualize_decision_boundary(reduced_X, Y, model, title='SVM'):
    '''
    Decision boundary 시각화 하는 함수
    '''
    t = np.array(Y)
    t = t.flatten()
    t = t.astype(np.integer)

    model.fit(reduced_X, t)

    plt.figure(figsize=[15, 10])
    plot_decision_regions(reduced_X,
                          t,
                          clf=model,
                          hide_spines=False,
                          colors='purple,limegreen',
                          markers=['^', 'v'])
    plt.title('{}'.format(title))
    plt.show()
```

```python
SVM = svm.SVC(kernel='linear')
SVM.fit(X_train, y_train)
prediction = SVM.predict(X_test)
print('SVM 정확도: ', metrics.accuracy_score(prediction, y_test))

visualize_decision_boundary(train_X_reduced, y_train, SVM, 'SVM')
```

### 이상치 처리

- 이상치를 하나도 허용하지 않는다면, 1. 결정 경계를 구할 수 없거나, 2. 과대적합 문제가 발생
- 따라서, 이상치를 어느정도 허용하는 **유연한 모델이 필요함!!!**
- **소프트 마진 분류**(적당한 오류는 허용) ↔ **하드 마진 분류**(모든 샘플을 오류 없이 분류)
    - **C: *float, default=1.0***
    - C는 규제의 정도(오류를 얼마나 허용할지)에 반비례\
        → C값이 클수록, 규제의 정도는 작아진다(오류를 적게 허용한다)

```python
hard_margin_SVM = svm.SVC(kernel='linear', C=9999)
hard_margin_SVM.fit(X_train, y_train)
prediction = hard_margin_SVM.predict(X_test)
print('Hard-margin-SVM 정확도: ', metrics.accuracy_score(prediction, y_test))

visualize_decision_boundary(train_X_reduced, y_train, hard_margin_SVM, 'Hard-margin SVM')
```

```python
soft_margin_SVM = svm.SVC(kernel='linear', C=0.01)
soft_margin_SVM.fit(X_train, y_train)
prediction = soft_margin_SVM.predict(X_test)
print('Soft-margin-SVM 정확도: ', metrics.accuracy_score(prediction, y_test))

visualize_decision_boundary(train_X_reduced, y_train, soft_margin_SVM, 'soft-margin SVM')
```

## Decision Tree



- 데이터에 있는 규칙을 학습해서 **분류 규칙**을 만들고, 새로 관측하는 데이터의 클래스 값을 찾는 알고리즘
- **‘예/아니오’의 답변을 요구**하는 질문을 반복하면서 데이터 분류
- 나무를 뒤집어 놓은 듯한 구조

### 용어 정리
- Decision Node(규칙 노드): 질문을 담은 네모상자
- Root Node(루트 노드): 첫 질문, 트리의 시작점, 가장 상단에 위
- Leaf Node(리프 노드): 자식이 없는 노드, 트리의 가장 하단에 위치
- Parent Node(부모 노드): 특정 노드에 연결된 이전 레벨의 노드
- Child Node(자식 노드): 특정 노드에 연결된 다음 레벨의 노드
- Sibling Node(형제 노드): 같은 부모를 가진 노드
- Level(레벨): 루트 노드를 기준으로 특정 노드까지의 경로 길이
- Depth(깊이): 트리의 최대 레벨
- Degree(차수): 특정 노드에 연결된 자식 노드의 수

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

#DecisionTree Classifier 생성
dt_clf=DecisionTreeClassifier(random_state=156)

#붗꽃 데이터를 로딩하고, 학습과 테스트 데이터 세트로 분리
iris_data=load_iris()
X_train,X_test,y_train,y_test=train_test_split(iris_data.data,iris_data.target,
                                              test_size=0.2,random_state=11)

#DecisionTressClassifer 학습
dt_clf.fit(X_train,y_train)

dt_clf.predict(X_test)
```

```python
# 결정트리 시각화
# graphviz를 쓰려면 anaconda prompt에서 conda install python-graphviz 를 입력해서 설치한 뒤, 주피터 노트북을 재실행한다.

import graphviz
from sklearn.tree import export_graphviz

a=export_graphviz(dt_clf, class_names=iris_data.target_names,
                   feature_names = iris_data.feature_names, impurity=True, filled=True)

test=open("test.txt",'w')
with open("test.txt", "w") as file:
    file.write(a)
test.close()

# 위에서 생성된 tree.dot 파일을 Graphviz가 읽어서 주피터 노트북상에서 시각화

with open("test.txt") as f:
    dot_graph=f.read()
graphviz.Source(dot_graph)
```

- 즉, 최대한 균일한 데이터 세트를 구성할 수 있도록 분할하자!!
    
        
    C > B > A 순서로 균일하다
    
### 균일도 측정 방법
- **엔트로피 계수(정보의 혼잡도)**
    - 부모 노드의 엔트로피 지수를 가장 많이 감소시키는 설명변수와 분리 값을 기준으로 자식 노드를 생성
    - 정보 이득 지수 = 1 - 엔트로피 계수
    
    
- **지니 계수**
    - 랜덤으로 선택한 데이터를 임의로 분류했을 때, 틀릴 확률
            
$$Gini = 1-\overset{C}{\underset{i=1}{\sum}}(p_i^2)$$
            


데이터가 혼잡할수록 지니계수는 커진다
----
정보 이득 지수가 높거나 지니 계수가 낮은 조건을 찾아서 데이터 분할


```python
gini_dt_clf=DecisionTreeClassifier(random_state=156, criterion='gini')
entropy_dt_clf=DecisionTreeClassifier(random_state=156, criterion='entropy')


#DecisionTressClassifer 학습
gini_dt_clf.fit(X_train,y_train)
entropy_dt_clf.fit(X_train,y_train)

print('기준이 지니계수일 때 결과:', gini_dt_clf.score(X_test, y_test))
print('기준이 엔트로피일 때 결과:', entropy_dt_clf.score(X_test, y_test))
```

    - 말단 노드(리프)의 최대 개수

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("2 feature, 3 classes Data")

#샘플 데이터 생성
X_features,y_labels=make_classification(n_features=2,n_redundant=0,n_informative=2,
                                       n_classes=3,n_clusters_per_class=1,random_state=0)


#그래프 형태로 2개의 피처로 2차원 좌표 시각화, 각 클래스의 값은 다른 색으로 표시됨

plt.scatter(X_features[:,0],X_features[:,1],marker='o',c=y_labels,s=25,edgecolor='k')
```

```python
import numpy as np

# Classifier의 Decision Boundary를 시각화 하는 함수
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # 학습 데이타 scatter plot으로 나타내기
    ax.scatter(X[:, 0], X[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
# 호출 파라미터로 들어온 training 데이타로 model 학습 . 
    model.fit(X, y)
    
# meshgrid 형태인 모든 좌표값으로 예측 수행. 
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
# contourf() 를 이용하여 class boundary 를 visualization 수행. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow', clim=(y.min(), y.max()),
                           zorder=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

#특정한 트리 생성 제약 없는 결정 트리(과대적합)의 학습과 결정 경계 시각화

dt_clf=DecisionTreeClassifier().fit(X_features,y_labels)
visualize_boundary(dt_clf,X_features,y_labels)
```

```python
#min_samples_leaf=4으로 트리 생성 조건을 제약한 결정 경계 시각화

print('min_samples_leaf=4일 때 결정경계')
dt_clf=DecisionTreeClassifier(min_samples_leaf=4).fit(X_features,y_labels)
visualize_boundary(dt_clf,X_features,y_labels)
```

```python
#max_depth=3으로 트리 생성 조건을 제약한 결정 경계 시각화

print('max_depth=3일 때 결정경계')
dt_clf_2=DecisionTreeClassifier(max_depth=3).fit(X_features,y_labels)
visualize_boundary(dt_clf,X_features,y_labels)
```

## KNN

__K-최소 근접(K-Nearest Neighbor) 알고리즘__   
KNN 알고리즘은 데이터들 간의 유사한 특징을 기준으로 분류하는 알고리즘이다. 주어진 데이터의 주변 이웃 데이터들의 특징을 파악하고 가장 유사한 데이터 그룹에 포함되도록 하는 방식이다.

```python
# 35마리의 도미 데이터
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]

# 14마리의 빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

length= bream_length + smelt_length
weight= bream_weight + smelt_weight

fish_data = [[l, w] for l, w in zip(length, weight)]

fish_target = [1]*35 + [0]*14 # Y데이터 
print(fish_target)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.xlabel("length")
plt.ylabel('weight')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier #KNN 분류 

kn=KNeighborsClassifier() #머신 on 

kn.fit(fish_data, fish_target) #물고기 데이터들을 X,Y에 넣어서 머신을 학습시켜서 규칙을 찾게 한다. 
```

```python

kn.score(fish_data, fish_target) # 정확도 측정. 즉 모델이 정확도가 100퍼센트라 도미와 빙어를 완벽하게 분류 가능해졌다. 
```

```python
kn.predict([[30,600]]) # 그래서 길이가 30, 무게가 600kg 그램으로 정해진 데이터를 넣어서 어느 팀에 들어가는지 알려주는 KNN
```

## 앙상블

- 여러 개의 분류기를 생성하고, 그 예측을 결합함으로써, 보다 정확한 최종 예측을 도출하는 기법
- 다양한 분류기의 예측 결과를 결합함으로써 단일 분류기보다 높은 예측값을 얻을 수 있음
- 프로젝트의 마지막에 다다르면 흔히 앙상블 방법을 사용해서 이미 만든 여러 괜찮은 예측기를 연결해서 더 좋은 예측기를 만듦

앙상블은 정형 데이터를 다루는데 가장 뛰어난 성능을 내는 알고리즘이다. 이 알고리즘은 대부분 결정 트리를 기반으로 만들고 있습니다.

### 랜덤 포레스트

랜덤 포레스트는 앙상블 학습의 대표 주자로 안정적인 성능 덕분에 널리 사용되고 있습니다. 이름 자체로 유추할 수 있듯이 랜덤 포레스트는 결정 트리를 랜덤하게 만들어 결정 트리의 숲을 만듭니다. 그리고 각 결정 트리의 예측을 사용해 최종 예측을 만듭니다. 

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
wine = pd.read_csv("https://bit.ly/wine_csv_data")
data= wine[['alcohol','sugar','pH']].to_numpy()
target = wine['class'].to_numpy()
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state= 42)
```

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=-1, random_state=42)
```

```python
scores = cross_validate(rf, X_train, y_train, return_train_score= True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
rf.fit(X_train, y_train)
print(rf.feature_importances_) #feature 이름: 알코올 도수, 당도, PH 
```

random forest의 장점은 특성의 일부를 랜덤하게 선택하여 결정 트리를 훈련합니다. 그 결과 하나의 특성에 과도하게 집중되지 않고 좀 더 많은 특성이 훈련에 기여할 기회를 얻으면서 과대적합을 줄이고 일반화 성능을 높이는데 도움이 됩니다.

### 그레이디언트 부스팅

그레디언트 부스팅은 깊이가 얕은 결정 트리를 이용하여 이전 트리의 오차를 보완하는 방식으로 앙상블하는 방법입니다.  
깊이가 얕은 결정 트리를 사용하기에 과대적합에 강하고 일반적으로 높은 일반화 성능을 기대할 수 있습니다. 

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
gb= GradientBoostingClassifier(random_state=42)
scores= cross_validate(gb, X_train, y_train, return_train_score = True, n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

여기서 알아야 할 점은 두개의 test 스코어와 train 스코어 두개의 수치가 그렇게 차이가 나지 않다는 점입니다. 그래서 학습이 편중되지 않고 과대 적합 과소 적합의 문제가 발생하지 않은 것을 알 수 있습니다.

```python
gb.fit(X_train, y_train)
print(gb.feature_importances_)
```
