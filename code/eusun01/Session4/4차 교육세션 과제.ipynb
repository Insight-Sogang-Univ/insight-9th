# 통계 사전과제

통계: 표본을 분석해서 모집단을 이해하는 활동
기술통계: 데이터를 정리, 요약하는 등 가지고 있는 데이터를 설명
추측통계: 과거의 데이터를 분석하여 미래의 모르는 데이터를 예측하는 추측

기술통계
- 수치 기술 통계: 숫자 형태로 데이터를 요약
  (중심 위치 척도, 변동성 척도, 연관성 척도)
표와 그래프 기술통계
- 표와 그래프 형태로 데이터 요약
- 데이터 시각화
  (도수분포표와 분할표, 막대그림, 파이그림. 히스토그램, 줄기-잎, 상자 그림 등등)

평균
- 모든 자료값의 합 / 전체 개수
- 이상치, 측정 오차에 민감하게 반응함

중앙값
- 자료를 크기 순으로 늘어 놓았을 때 중간에 위치하는 값

최빈값
- 자료 중 가장 빈도수가 높은 값

범위
- 최댓값 - 최솟값

분산
- 자료 각각이 평균으로부터 떨어져있는 정도
- E(편차제곱)

변동계수
- 표준편차를 평균에 대한 상대적인 값으로 표현한 것
- 표준편차/평균 * 100

공분산
- 두 변수가 각자의 평균으로부터 떨어진 값을 구하여 서로 곱하고 평균냄
- 두 변수가 어느 방향으로 얼마나 변동하는지를 나타냄
- 양수이면 두 변수는 양의 관계, 음수이면 음의 관계, 0이면 양도 음도 아닌 관계

상관계수
- -1과 1사이의 값을 가짐
- 공분산을 각 변수의 변동성을 이용하여 표준화한 값

확률분포
- 확률변수가 특정한 값을 가질 확률을 나타내는 함수

중요한 몇가지 확률 분포의 예시
- 정규분포
- T분포
- 카이스퀘어 분포
- F분포
- 이항분포
- 포아송 분포

모집단
- EX) 우리나라 성인 평균 키

표본집단
- EX) 추출된 집단

대수의 법칙(라플라스 정리)
- 표본집단의 크기가 커지면 그 표본평균이 모평균에 가까워짐

중심 극한 정리
- 무작위로 추출된 표본의 크기가 커질수록 표본 평균의 분포는 정규분포에 가까워짐
- 한번 추출할 때 많은 표본을 추출할수록 그 표본들의 평균 분포가 정규분포에 가까워짐

가설검정
- 절차: 가설 설정 -> 유의수준 설정 -> 검정통계량 산출 -> 기각/채택 판단

1종오류/2종오류
- 1종오류: 귀무가설이 실제로 참이었는데 기각할 확률
- 2종오류: ;귀무가설이 실제로 거짓인데 기각하지 못할 확률

조건부확률
- 어떤 사건이 일어났을 때, 그와 연관된 다른 사건이 일어날 확률
- P(B|A)로 주로 표시함(사건 A가 일어났을 때, 사건 B가 일어날 확률)

Bayes정리
- 조건부 확률을 이용하여 사전정보가 주어졌을 때 그 정보를 이용하여 새로운 정보를 추론하는 방법
- P(A|B) = P(B|A) * P(A) / P(B)와 같이 표현

```python
import numpy as np

fish_data= np.column_stack((fish_length, fish_weight))
fish_target= np.concatenate((np.ones(35), np.zeros(14)))
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state=42)

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

```python
distances, indexes = kn.kneighbors([[25,150]])

import matplotlib.pyplot as plt
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25,150,marker='^')
plt.scatter(train_input[indexes,0], train_input[indexes,1],marker='D')
plt.xlabel('length')
plt.ylabel('weight')
```
