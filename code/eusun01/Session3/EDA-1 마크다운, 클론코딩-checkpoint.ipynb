Feature의 종류

데이터의 성질을 파악한 후 그에 맞는 플롯 사용
- 시간에 따른 변화 -> line, area, bar
- 비교와 랭킹 -> bar

- 분포 -> box plot, histogram
- 부분이 전체에 차지하는 정도 -> pie, bar

Matplotlib
- EDA에서 일반적으로 사용
- 파이썬 언어 및 넘피 라이브러리를 활용함

Seaborn 
- Matplotlib을 기반으로 만들어져 통계 데이터 시각화에 최적화된 라이브러리

수치형 데이터 시각화
히스토그램
- 수치형 데이터의 구간별 빈도수를 나타낸 그래프
- 수치형 데이터 하나에 대한 빈도를 나타냄
- 특정 범주별로도 구분 가능

커널밀도추정 함수 그래프
- 히스토그램을 매끄럽게 곡선으로 연결한 그래프

막대그래프
- 범주형 데이터 값에 따라 수치형 데이터 값이 어떻게 달라지는지 파악할때 사용

포인트플롯
- 막대 그래프와 동일한 정보 제공(모양만 다름)

박스플롯
- 5가지 요약 수치 제공
- 막대그래프, 포인트플롯보다 더 많은 정보 제공
- 제 1사분위 수(Q1) : 전체 데이터 중 하위 25%에 해당하는 값
- 제 2사분위 수(Q2) : 50%에 해당하는 값(중앙값)
- 제 3사분위 수(Q3) : 상위 25%에 해당하는 값

바이올린 플롯
- 박스 플롯 + 커널밀도추정 함수 그래프

카운트플롯

파이그래프

# 클론코딩

```python
import seaborn as sns
data = sns.load_dataset("titanic")
display(data.head())
```

```python
# 히스토그램
sns.histplot(x='age',data = data) 
```

```python
sns.histplot(x='age', hue = 'alive', data=data)
```

```python
# 커널밀도추정 함수 그래프
sns.kdeplot(x='age', data=data)
```

```python
# 막대그래프
sns.barplot(x='class',y='fare', data=data)
```

```python
# 포인트플롯
sns.pointplot(x='class', y='fare', data=data)
```

```python
# 박스플롯
sns.boxplot(x='class', y='age', data=data)
```

```python
# 바이올린 플롯
sns.violinplot(x='class', y='age', data=data)
```

```python
# 바이올린 플롯-성별에 따른 등급별 나이 분포 보기
sns.violinplot(x='class', y='age', hue='sex', data=data)
```

```python
# 데이터 합치기
sns.violinplot(x='class', y='age', hue='sex', data=data, split=True)
```

```python
# 카운트플롯
sns.countplot(x='class', data=data)
```

```python
#카운트플롯
sns.countplot(y='class', data=data)
```

```python
# 파이그래프
import matplotlib.pyplot as plt

x=[10,60,30]
labels=['A','B','C']

plt.pie(x, labels=labels, autopct='%.1f%%')
```
