# 주피터 노트북

1. 인터랙티브 컴퓨팅을 위한 서비스. 여러 개 언어를 통한 실행환경을 지원하며 핵심 언어는 Julia, Python, R이다.

2. 셀이 'run'의 단위가 되는데, 이는 '.py' 확장자의 파이썬 파일이 실행되는 방식과 다르다. '.py'로 실행할 경우 처음부터 끝까지가 실행범위가 되지만, Jupyter Notebook 상에서 '.ipynb' 파일을 셀 단위로 실행하는 경우 하나의 셀이 실행범위가 된다. Python을 "대화식"으로 실행하는 것과 동일한 것으로 이해할 수 있다.
    * 셀을 통한 실행에서 print()를 호출하지 않고도 변수의 값을 확인 및 출력할 수 있다.

3. 셀은 markdown 타입과 code 타입이 있는데, markdown을 활용해 가시성 높은 문서를 작성할 수 있으며, 데이터 시각화와 함께 중요한 시각적 요소 중 하나이다.

4. 출력결과가 파일에 저장되기에 다른 사용자가 코드를 실행하지 않아도 결과를 확인하 수 있는 장점이 있다. '.ipynb' 파일은 데이터분석을 실행하는 코드이면서 그 결과를 보여주는 문서로 이해할 수 있다.

# Git
1. "분산버전관리시스템"
2. 버전관리란, 과거 데이터에 대한 백업을 의미한다.
3. 이전 버전의 코드를 백업할 필요성으로 사용한다.
4. "과거 상태에 대한 저장"
5. 명령어 정리
    * git init : 로컬에 저장소를 만든다
    * git add : 해당 파일을 추적하도록 한다(도마!)
    * git status : git의 현재 상태
    * git commit -m "메세지" : 버전을 생성한다
    * git push : 원격 저장소로 push 한다
    * git log -p : 지금까지의 변경사항을 커밋 메세지와 함께 조회한다
    * git diff : git add 전과 후를 비교
6. 원격저장소
    * 파일의 업로드/다운로드는 "싱크/버전을 맞추는" 동기화로서 이해할 수 있다.
    * 업로드(push)는 원격(remote) 저장소의 파일을 로컬(local)의 파일로 동기화하는 것이며,
    * 다운로드(pull)은 로컬(local)에 있는 파일을 원격(remote) 저장소의 파일로 동기화하는 것으로 이해할 수 있다.
    * git push, git pull
