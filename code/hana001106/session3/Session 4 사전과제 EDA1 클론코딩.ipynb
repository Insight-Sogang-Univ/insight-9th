```python
import seaborn as sns # seaborn 라이브러리를 쓰기 위함

data = sns.load_dataset("titanic") # titanic data 불러오기
display(data.head())
```

```python
# 히스토그램
sns.histplot(x='age',  data=data)
```

```python
# 히스토그램에 특정 범주 추가하기
sns.histplot(x='age', hue = 'alive', data = data)
```

```python
# 커널 밀도 추정 함수
sns.kdeplot(x = "age", data = data)
```

```python
# barplot
sns.barplot(x='class', y='fare', data = data)
```

```python
# pointplot 막대 그래프와 비슷함
sns.pointplot(x = 'class', y = 'fare', data = data)
```

```python
# boxplot 사분위수 및 이상치 제공
sns.boxplot(x = 'class', y = 'age', data = data)
```

```python
# violinplot boxplot+kdeplot
sns.violinplot(x = "class", y = "age", data = data)
```

```python
# 성별별로 나누기
sns.violinplot(x='class', y='age', hue='sex', data = data)
```

```python
# 성별별로 나눈 걸 합쳐서 보기
sns.violinplot(x='class', y='age', hue='sex', data=data, split=True)
```

```python
# count plot
sns.countplot(x='class', data=data)
```

```python
sns.countplot(y='class', data=data) # 위에 거와 차이는 x축 방향이냐 y축 방향이냐
```

```python
# pie graph
import matplotlib.pyplot as plt # pieplot은 pyplot 불러오기

x = [10, 60, 30]
labels = ['A', 'B', 'C']

plt.pie(x, labels = labels, autopct='%.1f%%')
```

##### (추가사항) 위에 있는 그래프들을 그리면 그래프 위에 이상하게 영어로 뭐가 많이 뜨는데 이를 없애기 위해서는 plt.show()을 사용하면 됨

```python
x = [10, 60, 30]
labels = ['A', 'B', 'C']

plt.pie(x, labels = labels, autopct='%.1f%%')
plt.show()
```
