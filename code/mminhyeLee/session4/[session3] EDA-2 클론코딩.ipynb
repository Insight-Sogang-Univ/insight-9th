#  EDA란?
- EDA(Exploratory Data Analysis, 탐색적 데이터분석)란, 데이터의 특징과 내재하는 구조적인 관계를 알아내기 위한 분석 기법
- EDA를 통해 수집한 데이터를 다양한 각도에서 관찰하고 이해할 수 있음
- 데이터를 보다 직관적으로 바라보기 위해 *통계적인 방법*과 *데이터 시각화*가 사용됨

**Tabular data(정형 데이터) 유형의 Binary Classification(이진 분류) 문제인 "Titanic"을 활용**

- 데이터 유형에 따라 다양한 시각화 방식 활용
- 결측치 처리
3. Feature Engineering과 Data Cleaning
- Feature 분석에서 얻은 insight로 새로운 feature 만들기
- machine learning을 위한 cleaning

## Feature Engineering & Data Cleaning
- Feature Engineering: 초기 데이터의 특징을 가공하고 생산하여 모델 입력 데이터를 생성하는 과정
    - ex) Name feature를 활용하여 Initial feature 생성
    - 문자보다는 숫자, 개별값보다는 범위로 데이터 가공

```python
# Age feature 가공: Age_Band

data['Age_band'] = 0
data.loc[data['Age']<=16, 'Age_band'] = 0
data.loc[(data['Age']>16) & (data['Age']<=32), 'Age_band'] = 1
data.loc[(data['Age']>32) & (data['Age']<=48), 'Age_band'] = 2
data.loc[(data['Age']>48) & (data['Age']<=64), 'Age_band'] = 3
data.loc[(data['Age']>64), 'Age_band'] = 4

data.head()
```

```python
data['Age_band'].value_counts().to_frame().style.background_gradient(cmap='summer_r')
```

```python
sns.factorplot('Age_band', 'Survived', col='Pclass', data=data)
plt.show()
```

```python
# Family_Size and Alone : Parch 와 SibSp feature를 합쳐 새 feature 생성
data['Family_Size'] = 0
data['Family_Size'] = data['Parch'] + data['SibSp']
data['Alone'] = 0
data.loc[data["Family_Size"]==0, 'Alone'] = 1

sns.factorplot('Family_Size', 'Survived', data=data, ax=ax[0])
ax[0].set_title("Family_Size vs Survived")

sns.factorplot('Alone', 'Survived', data=data, ax=ax[1])
ax[1].set_title('Alone vs Survived')

plt.show()
```

```python
sns.factorplot('Alone', 'Survived', col='Pclass', hue='Sex', data=data)
plt.show()
```

```python
# Fare_Range
data['Fare_Range']=pd.qcut(data['Fare'], 4) # Fare feature를 4등분!
data.groupby('Fare_Range')['Survived'].mean().to_frame().style.background_gradient(cmap='summer_r')
```

```python
# Fare_cat: Fare_Range 값을 singleton 값으로 바꾸기

data['Fare_cat'] = 0
data.loc[data['Fare']<= 7.91, 'Fare_cat'] = 0
data.loc[(data['Fare']>7.91) & (data['Fare']<=14.454), 'Fare_cat'] = 1
data.loc[(data['Fare']>14.454) & (data['Fare']<=31), 'Fare_cat'] = 2
data.loc[data['Fare']>31, 'Fare_cat'] = 3

sns.factorplot('Fare_cat', 'Survived', data=data, hue='Sex')
plt.show()
```

```python
# convert string values into Numeric

data['Sex'].replace(['male', 'female'], [0, 1], inplace=True)
data['Embarked'].replace(['S', 'C', 'Q'], [0, 1, 2], inplace=True)
data['Initial'].replace(['Mr', 'Mrs', 'Miss', 'Master', 'Other'], [0,1, 2, 3, 4],
                       inplace=True)
data.head()
```

```python
# dropping unneeded features
data.columns
```

```python
data.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin', 'Fare_Range', 'Age'],
         axis=1, inplace = True)
sns.heatmap(data.corr(), annot=True, cmap='RdYlGn', linewidths=0.2, annot_kws={'size':20})

fig = plt.gcf()
fig.set_size_inches(18, 15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
